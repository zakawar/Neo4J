```python
from py2neo import Graph
import pprint
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
graph = Graph(os.getenv("Connection_URI"), auth=(os.getenv("USER"), os.getenv("PASSWORD")))
```

### Return the 10 first Disease's nodes

```python
query1 = "MATCH (n:Disease) RETURN n LIMIT 10"
result = graph.run(query1).data()
pprint.pprint(result)
```

### Return the 10 first Gene's nodes

```python
query2 = "MATCH (n:Gene) RETURN n LIMIT 10"
result = graph.run(query2).data()
pprint.pprint(result)
```

### Return the 5 first nodes where the disease's name start with 'activated'

```python
query3 = "MATCH (d:Disease)-[p]-(g:Gene) \
WHERE d.AName STARTS WITH 'activated' \
RETURN d.AName,p,g.BName LIMIT 5"

result = graph.run(query3).data()
pprint.pprint(result)
```

## Running query in the Console


### Return the 20 first nodes where the disease's name start with subword 'ac' (i.e. acanthosis, actived , acne,... )

```python
query4 = "MATCH (d:Disease)-[p]-(g:Gene) WHERE d.AName =~ 'ac.*' RETURN d.AName,p,g.BName limit 20"
result = graph.run(query4).data()
pprint.pprint(result)
```
